```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
import numpy as np

# Load dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize
X_train = X_train / 255.0
X_test = X_test / 255.0

# Flatten
X_train = X_train.reshape(-1, 784)
X_test = X_test.reshape(-1, 784)

# One-hot encoding
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Model
model = Sequential([
    Dense(10, activation='softmax', input_shape=(784,))
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.summary()

model.fit(X_train, y_train, epochs=10, batch_size=64)

test_loss, test_acc = model.evaluate(X_test, y_test)
print("Test accuracy:", test_acc)
```

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt

# -----------------------------
# Load dataset
# -----------------------------
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize
X_train = X_train / 255.0
X_test = X_test / 255.0

# Flatten
X_train = X_train.reshape(-1, 784)
X_test = X_test.reshape(-1, 784)

# One-hot encoding
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Batch sizes to compare
batch_sizes = [32, 64, 128, 256]

histories = {}

# -----------------------------
# Train model for each batch size
# -----------------------------
for batch in batch_sizes:

    print(f"\nTraining with batch size = {batch}\n")

    model = Sequential([
        Dense(10, activation='softmax', input_shape=(784,))
    ])

    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    history = model.fit(
        X_train,
        y_train,
        epochs=10,
        batch_size=batch,
        validation_data=(X_test, y_test),
        verbose=1
    )

    histories[batch] = history


# -----------------------------
# Plot Accuracy Comparison
# -----------------------------
plt.figure()

for batch in batch_sizes:
    plt.plot(histories[batch].history['val_accuracy'], label=f'Batch {batch}')

plt.title("Validation Accuracy Comparison")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()

plt.show()


# -----------------------------
# Plot Loss Comparison
# -----------------------------
plt.figure()

for batch in batch_sizes:
    plt.plot(histories[batch].history['val_loss'], label=f'Batch {batch}')

plt.title("Validation Loss Comparison")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()

plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
import numpy as np

# -----------------------------
# Load dataset
# -----------------------------
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize
X_train = X_train / 255.0
X_test = X_test / 255.0

# Flatten
X_train = X_train.reshape(-1, 784)
X_test = X_test.reshape(-1, 784)

# One-hot encoding
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Epoch values to test
epoch_list = [10, 15, 20]

# Dictionary to store results
histories = {}

# -----------------------------
# Training models
# -----------------------------
for ep in epoch_list:

    print(f"\nTraining model with epochs = {ep}\n")

    model = Sequential([
        Dense(10, activation='softmax', input_shape=(784,))
    ])

    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    model.summary()

    history = model.fit(
        X_train,
        y_train,
        epochs=ep,
        batch_size=64,
        validation_data=(X_test, y_test),
        verbose=1
    )

    histories[ep] = history

    test_loss, test_acc = model.evaluate(X_test, y_test)
    print("Test accuracy:", test_acc)


# -----------------------------
# Accuracy Graph
# -----------------------------
plt.figure()

for ep in epoch_list:
    plt.plot(histories[ep].history['val_accuracy'], label=f'Epochs {ep}')

max_epoch = max(epoch_list)

plt.title("Validation Accuracy Comparison")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")

# X-axis as whole numbers
plt.xticks(np.arange(1, max_epoch + 1, 1))

plt.legend()
plt.show()


# -----------------------------
# Loss Graph
# -----------------------------
plt.figure()

for ep in epoch_list:
    plt.plot(histories[ep].history['val_loss'], label=f'Epochs {ep}')

plt.title("Validation Loss Comparison")
plt.xlabel("Epoch")
plt.ylabel("Loss")

# X-axis as whole numbers
plt.xticks(np.arange(1, max_epoch + 1, 1))

plt.legend()
plt.show()
```

Added Dense Layer-32,64,128, Activation Function-RELU And For o/p-10, Activation Function-Softmax

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize
X_train = X_train / 255.0
X_test = X_test / 255.0

# Flatten
X_train = X_train.reshape(-1, 784)
X_test = X_test.reshape(-1, 784)

# One-hot encoding
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)


def train_and_plot(model, title):

    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    print("\n", title)
    model.summary()

    history = model.fit(
        X_train, y_train,
        epochs=10,
        batch_size=64,
        validation_data=(X_test, y_test),
        verbose=1
    )

    test_loss, test_acc = model.evaluate(X_test, y_test)
    print("Test accuracy:", test_acc)

    # Accuracy curve
    plt.figure()
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title(title + " - Accuracy")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.show()

    # Loss curve
    plt.figure()
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(title + " - Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()
    plt.show()


# ------------------------------------------------
# Model 1: Add Dense layer with 32 neurons
# ------------------------------------------------
model1 = Sequential([
    Dense(32, activation='relu', input_shape=(784,)),
    Dense(10, activation='softmax')
])

train_and_plot(model1, "Model with 32 Neurons")


# ------------------------------------------------
# Model 2: Add Dense layer with 64 neurons
# ------------------------------------------------
model2 = Sequential([
    Dense(32, activation='relu', input_shape=(784,)),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

train_and_plot(model2, "Model with 32 + 64 Neurons")


# ------------------------------------------------
# Model 3: Add Dense layer with 128 neurons
# ------------------------------------------------
model3 = Sequential([
    Dense(32, activation='relu', input_shape=(784,)),
    Dense(64, activation='relu'),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

train_and_plot(model3, "Model with 32 + 64 + 128 Neurons")
```
